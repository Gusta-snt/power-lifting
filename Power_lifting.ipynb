```python
import pandas as pd

power_lifting = pd.read_csv("powerlifting.csv")

power_lifting
```

```python
exercicios = power_lifting["Lift Type"].unique().tolist()
```

```python
for exercicio in exercicios:
  media = power_lifting.loc[power_lifting["Lift Type"] == exercicio, "Amount Lifted (kg)"].mean()
  print(f"O exercÃ­cio {exercicio} tem mÃ©dia de {round(media, 2)} kg")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.scatterplot(x='Age', y='Amount Lifted (kg)', data=power_lifting)
plt.show()
```

```python
classes = power_lifting["Weight Class"].unique().tolist()

maior = 0
classe_maior = ""
for classe in classes:
  media = power_lifting.loc[power_lifting["Weight Class"] == classe, "Amount Lifted (kg)"].mean()
  if media > maior:
    maior = media
    classe_maior = classe

print(f"A classe que levanta mais quilos em mÃ©dia Ã© {classe_maior}, levantando {round(float(maior), 2)} kg")
```

```python
nomes = power_lifting["Lifter Name"].unique().tolist()
medias = []
for nome in nomes:
  media = power_lifting.loc[power_lifting["Lifter Name"] == nome, "Amount Lifted (kg)"].mean()
  medias.append(media)
  print(f"O levantador {nome} tem mÃ©dia de {round(media, 2)} kg")
```

```python
plt.bar(nomes, medias)
plt.xticks(rotation=45, ha="right")
plt.show()
```
